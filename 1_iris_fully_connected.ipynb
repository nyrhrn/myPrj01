```python
# Import useful libraries
import numpy as np
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Load dataset
from sklearn.datasets import load_iris
iris_data=load_iris()
iris_data["target"]
# iris_data["target"].reshape(-1,1) # 
# iris_data.target = iris_data["target"] # Convert data to a singlr column
x=iris_data["data"]
y=tf.keras.utils.to_categorical(iris_data["target"].reshape(-1,1),num_classes=3)
```

```python
# Split the data for training and testing
train_x, test_x ,train_y , test_y=train_test_split(x,y,test_size=0.20)
```

```python
# Build the model
model=Sequential()
model.add(Dense(10,activation='relu',input_dim=4))
model.add(Dense(10,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
# Adam optimizer 
optimizer=tf.keras.optimizers.Adam(learning_rate=0.001)
# Compile model
model.compile(optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train the model
model.fit(train_x, train_y, batch_size=5,epochs=200)
```

```python
# Test on unseen data
result=model.evaluate(test_x,test_y)
print(f'Final test set loss :{result[0]}')
print('--------------------------------')
print(result[0])
print("{0:.4f}".format(0.1234567890))
print('Final test set loss : {:.4f}'.format(result[0]))
print('Final test set accuracy : {:.4f}'.format(result[1]))
```
